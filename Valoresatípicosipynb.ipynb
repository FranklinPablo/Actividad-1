```python
data1=pd.read_csv('Inmuebles_sin_nulos.csv')
data1.head(10)
```

```python
data=data1.drop(['Unnamed: 0'], axis=1)
data
```

```python
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
fig=plt.figure(figsize=(10,5))
plt.hist(x=data["Precio Venta"],color='Orange',rwidth=0.80)
plt.title('Histograma de Precio de ventas')
plt.xlabel('Precio Venta')
plt.ylabel('Frecuencia')

plt.show()
```

```python
fig=plt.figure(figsize=(5,3))
plt.boxplot(data["Precio Venta"])
plt.title("Outliers de precio Venta")
plt.show()
```

```python
y=data["Precio Venta"]
limite_superior=y.mean() + 3*y.std()
limite_inferior=y.mean() - 3*y.std()
print("Límite superior permitido", limite_superior)
print("Límite inferior permitido", limite_inferior)
```

```python
outliers=data[(y>limite_superior)|(y<limite_inferior)]
outliers
```

```python
#datos limpios
data_clean= data[(y<=limite_superior)&(y>=limite_inferior)]
data_clean
```

```python
fig=plt.figure(figsize=(5,3))
plt.boxplot(data_clean["Precio Venta"])
plt.title("Outliers de precio Venta")
plt.show()
```

Método por inter quartiles

```python
y=data['Precio Venta']


percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

limite_superior_iqr= percentile75 + 1.5*iqr
limite_inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", limite_superior_iqr)
print("Limite inferior permitido", limite_inferior_iqr)
```

```python
outliers_iqr=data[(y>limite_superior_iqr)|(y<limite_inferior_iqr)]
outliers_iqr
```

```python
data_clean_iqr =data[(y<=limite_superior_iqr)&(y>=limite_inferior_iqr)]
data_clean_iqr
```

```python
fig=plt.figure(figsize=(5,3))
plt.boxplot(data_clean_iqr["Precio Venta"])
plt.title("Outliers de Precio venta")
plt.show()
```
